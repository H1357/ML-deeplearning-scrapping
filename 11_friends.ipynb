```python
def write_to_friends (LLL):
    conn = sqlite3.connect(dbase_friends)
    #print ("Opened database successfully")
    cursor = conn.cursor()

    cursor.executemany("INSERT INTO F VALUES (?,?)", LLL)

    conn.commit()
    #print ("Records created successfully")
    conn.close()
```
